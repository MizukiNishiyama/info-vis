```python
import lyricsgenius


# Genius APIのトークンを設定
token = '7q6YOSMqusnmi26X2hINB0izURIPqqNXKwO02daE0ZBmtZLLh-TOxs0roQSZg16N'


# LyricsGeniusクライアントを作成
genius = lyricsgenius.Genius(token)


# 検索したいアーティスト、楽曲名を設定
artist_name = 'Estelle'
track = 'American Boy'


# アーティストを検索
artist = genius.search_artist(artist_name, max_songs=5)


# 指定した曲を検索して、該当する曲の歌詞を取得
song_lyrics = genius.search_song(track, artist.name).lyrics


print(song_lyrics)
```
